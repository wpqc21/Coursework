# 1-Dimensional Random Walk

### 1 Walk, 10 Steps

```python
step = 0

random_walk = [0]

for x in range(10):
    step = random_walk[-1]
    walk = np.random.choice([-1, 1])
    step += walk
    random_walk.append(step)

print(random_walk)
```

```python
plt.plot(random_walk)
```

# 3-Dimensional Random Walk

### 5 Walks, 10 Steps

```python
step_x_a, step_y_a, step_z_a = 0, 0, 0
step_x_b, step_y_b, step_z_b = 0, 0, 0
step_x_c, step_y_c, step_z_c = 0, 0, 0
step_x_d, step_y_d, step_z_d = 0, 0, 0
step_x_e, step_y_e, step_z_e = 0, 0, 0

random_walk_x_a, random_walk_y_a, random_walk_z_a = [0], [0], [0]
random_walk_x_b, random_walk_y_b, random_walk_z_b = [0], [0], [0]
random_walk_x_c, random_walk_y_c, random_walk_z_c = [0], [0], [0]
random_walk_x_d, random_walk_y_d, random_walk_z_d = [0], [0], [0]
random_walk_x_e, random_walk_y_e, random_walk_z_e = [0], [0], [0]

for x in range(10):
    step_x_a, step_y_a, step_z_a = random_walk_x_a[-1], random_walk_y_a[-1], random_walk_z_a[-1]
    walk_x_a, walk_y_a, walk_z_a  = np.random.choice([-1, 1]), np.random.choice([-1, 1]), np.random.choice([-1, 1])
    step_x_a, step_y_a, step_z_a  = step_x_a + walk_x_a, step_y_a + walk_y_a, step_z_a + walk_z_a
    random_walk_x_a.append(step_x_a), random_walk_y_a.append(step_y_a), random_walk_z_a.append(step_z_a)
    
    step_x_b, step_y_b, step_z_b = random_walk_x_b[-1], random_walk_y_b[-1], random_walk_z_b[-1]
    walk_x_b, walk_y_b, walk_z_b  = np.random.choice([-1, 1]), np.random.choice([-1, 1]), np.random.choice([-1, 1])
    step_x_b, step_y_b, step_z_b  = step_x_b + walk_x_b, step_y_b + walk_y_b, step_z_b + walk_z_b
    random_walk_x_b.append(step_x_b), random_walk_y_b.append(step_y_b), random_walk_z_b.append(step_z_b)
    
    step_x_c, step_y_c, step_z_c = random_walk_x_c[-1], random_walk_y_c[-1], random_walk_z_c[-1]
    walk_x_c, walk_y_c, walk_z_c  = np.random.choice([-1, 1]), np.random.choice([-1, 1]), np.random.choice([-1, 1])
    step_x_c, step_y_c, step_z_c  = step_x_c + walk_x_c, step_y_c + walk_y_c, step_z_c + walk_z_c
    random_walk_x_c.append(step_x_c), random_walk_y_c.append(step_y_c), random_walk_z_c.append(step_z_c)
    
    step_x_d, step_y_d, step_z_d = random_walk_x_d[-1], random_walk_y_d[-1], random_walk_z_d[-1]
    walk_x_d, walk_y_d, walk_z_d  = np.random.choice([-1, 1]), np.random.choice([-1, 1]), np.random.choice([-1, 1])
    step_x_d, step_y_d, step_z_d  = step_x_d + walk_x_d, step_y_d + walk_y_d, step_z_d + walk_z_d
    random_walk_x_d.append(step_x_d), random_walk_y_d.append(step_y_d), random_walk_z_d.append(step_z_d)
    
    step_x_e, step_y_e, step_z_e = random_walk_x_e[-1], random_walk_y_e[-1], random_walk_z_e[-1]
    walk_x_e, walk_y_e, walk_z_e  = np.random.choice([-1, 1]), np.random.choice([-1, 1]), np.random.choice([-1, 1])
    step_x_e, step_y_e, step_z_e  = step_x_e + walk_x_e, step_y_e + walk_y_e, step_z_e + walk_z_e
    random_walk_x_e.append(step_x_e), random_walk_y_e.append(step_y_e), random_walk_z_e.append(step_z_e)

data1 = pd.DataFrame({ "X": random_walk_x_a, "Y": random_walk_y_a, "Z": random_walk_z_a, "Type": "a"})

data2 = pd.DataFrame({ "X": random_walk_x_b, "Y": random_walk_y_b, "Z": random_walk_z_b, "Type": "b"})

data3 = pd.DataFrame({ "X": random_walk_x_c, "Y": random_walk_y_c, "Z": random_walk_z_c, "Type": "c"})

data4 = pd.DataFrame({ "X": random_walk_x_d, "Y": random_walk_y_d, "Z": random_walk_z_d, "Type": "d"})

data5 = pd.DataFrame({ "X": random_walk_x_e, "Y": random_walk_y_e, "Z": random_walk_z_e, "Type": "e"})

random_walk_3d_10 = pd.concat([data1, data2, data3, data4, data5])

fig3D_10 = px.line_3d(random_walk_3d_10, x='X', y='Y', z='Z', color='Type')
# fig3D_10.show()

fig3D_10.write_image("fig3D_10.png")

fig3D_10_r = Image.open("fig3D_10.png")

fig3D_10_r
```
